## Question 2: 
#### Use Car_Purchasing_Data.csv file, this file is in the Linear Regression folder. In the auto example, find the multi-linear relationship between 'Age','Annual Salary', 'Net Worth' with 'Car Purchase Amount'. 

```python
auto = pd.read_csv('LinearReg/Car_Purchasing_Data.csv')
```

```python
auto.head()
```

```python
print(auto.columns)
print(auto.dtypes)
```

```python
# Dropping object columns

auto.drop(["Customer Name", "Customer e-mail", "Country"], \
          axis=1, inplace=True)
print(auto.shape)

print(auto.columns)
```

```python
# creating an instance of LinearRegression class
reg = linear_model.LinearRegression()
print(reg)
```

```python
auto.describe()
```

```python
auto.drop(['Gender', 'Credit Card Debt'], axis=1, inplace=True)
```

```python
auto.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a_scaler = StandardScaler() # creating an instance of Standard Scaler
```

```python
auto_independent = a_scaler.fit_transform(auto[['Age', 'Annual Salary', 'Net Worth']])
```

```python
print(type(auto_independent))
```

```python
auto[['Age', 'Annual Salary', 'Net Worth']].head()
```

```python
# after scaling
auto_independent[0:5]
```

```python
# using train_test_split(), we are splitting the data into training and test 
# x_train, x_test, y_train, y_test

x_train, x_test, y_train, y_test = train_test_split(auto_independent, auto["Car Purchase Amount"], test_size=0.2, random_state=4)
```

```python
# we have to fit the data
reg.fit(x_train, y_train)
```

```python
print(reg.coef_) # 
print(reg.intercept_)
```

```python
yhat = reg.predict(x_test) 
```

### Mean Squared Error for test set

```python
from sklearn.metrics import mean_squared_error

mse_test = mean_squared_error(y_test, yhat)
print(mse_test)
```

### Mean Squared Error for the train set

```python
# y_t_predict is the predicted y values for the x_train data
y_t_predict = reg.predict(x_train)

# note that y_train is the true y value
mse_train = mean_squared_error(y_train, y_t_predict)
print(mse_train)
```

```python
r1 = mse_test/mse_train

diff1 = np.abs((mse_train - mse_test)/mse_train)

print(r1)

print(diff1)
```

### r-squared for test and train set

```python
from sklearn.metrics import r2_score

print("r-squared for the test data: ", r2_score(y_test, yhat))
    
print("r-squared for the train data: ", r2_score(y_train, y_t_predict))
```

#### Conclusion:

r-squared for test is 0.9994  and for train is 0.9995

The multi-linear model seems like a good fit.
