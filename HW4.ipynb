```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
country_df = pd.read_csv('Documents/KStateSpring2025/MLAS390/Country-data.csv')

print(country_df.shape)
print(country_df.head())
```

```python
country_df.describe()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

import plotly.express as px
from collections import Counter
```

```python
X = country_df[["child_mort", "inflation"]]
```

```python
# Creating an instance of DBScan with eps = 15 and min_samples = 15
dbs = DBSCAN(eps=15, min_samples=15, metric="l2")

# predict
y_pred = dbs.fit_predict(X)

# Cluster numbers

country_df['Cluster Number'] = y_pred
```

```python
country_df['Cluster Number'].unique()
```

```python
# Number of points in each cluster

print(country_df['Cluster Number'].value_counts())
```

```python
px.histogram(country_df, country_df['Cluster Number'], color="Cluster Number")
```

```python
# Creating an instance of DBScan with eps = 5 and min_samples = 5
dbs = DBSCAN(eps=5, min_samples=5, metric="l2")

# predict
y_pred = dbs.fit_predict(X)

# Cluster numbers

country_df['Cluster Number'] = y_pred
```

```python
country_df['Cluster Number'].unique()
```

```python
# Number of points in each cluster

print(country_df['Cluster Number'].value_counts())
```

```python
px.histogram(country_df, country_df['Cluster Number'], color="Cluster Number")
```
